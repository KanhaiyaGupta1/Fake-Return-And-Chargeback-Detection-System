```python
import pymysql
import pandas as pd

# 1Ô∏è‚É£ Connect to MySQL
try:
    conn = pymysql.connect(
        host="localhost",
        user="root",
        password="Ayush@4618",
        database="fraud_detection"
    )
    cursor = conn.cursor()
    print("‚úÖ Connected to MySQL successfully!")
except Exception as err:
    print(f"‚ùå Error connecting to MySQL: {err}")

# 2Ô∏è‚É£ Input Transaction Details
customer_name = input("Enter customer name: ")
amount = float(input("Enter order amount: "))
previous_returns = int(input("Enter number of previous returns: "))
chargeback_history = int(input("Any previous chargebacks? (1=Yes / 0=No): "))
address_mismatch = int(input("Is address mismatched? (1=Yes / 0=No): "))
high_risk_country = int(input("From high-risk country? (1=Yes / 0=No): "))

# 3Ô∏è‚É£ Calculate Risk Score
risk_score = 0
if amount > 1000:
    risk_score += 2
if previous_returns > 2:
    risk_score += 2
if chargeback_history == 1:
    risk_score += 3
if address_mismatch == 1:
    risk_score += 2
if high_risk_country == 1:
    risk_score += 2

if risk_score >= 6:
    risk_level = "High"
elif risk_score >= 3:
    risk_level = "Medium"
else:
    risk_level = "Low"

print(f"\nRisk Score: {risk_score}")
print(f"Risk Level: {risk_level}")

# 4Ô∏è‚É£ Save Data to Database
try:
    query = """
    INSERT INTO transactions
    (customer_name, amount, previous_returns, chargeback_history,
     address_mismatch, high_risk_country, risk_score, risk_level)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
    """
    values = (customer_name, amount, previous_returns,
              chargeback_history, address_mismatch,
              high_risk_country, risk_score, risk_level)
    cursor.execute(query, values)
    conn.commit()
    print("\n‚úÖ Transaction saved successfully!")
except Exception as err:
    print(f"‚ùå Database Error: {err}")

# 5Ô∏è‚É£ Fetch and Show Data
try:
    query = "SELECT * FROM transactions"
    cursor.execute(query)
    rows = cursor.fetchall()
    df = pd.DataFrame(rows, columns=[desc[0] for desc in cursor.description])
    print("\nüìä Transaction Records:")
    display(df)
except Exception as e:
    print(f"‚ö† Could not fetch data: {e}")

# 6Ô∏è‚É£ Close Connection
cursor.close()
conn.close()
print("\nüîí MySQL connection closed.")
```
